```python
from PIL import Image
import os

# Define input and output directories
input_directory = r'C:\Users\Harsh\Downloads\Compressed\train'
output_directory = os.path.join(input_directory, 'comp')

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

new_size = (256, 256)
count = 0
batch_size = 1000

# Iterate through each file in the input directory
for filename in os.listdir(input_directory):
    if filename.endswith(".jpeg"):  # You can add more image extensions if needed
        input_image_path = os.path.join(input_directory, filename)
        output_image_path = os.path.join(output_directory, filename)
        try:
            with Image.open(input_image_path) as img:
                img = img.resize(new_size)
                img.save(output_image_path)
                count += 1
                if count % batch_size == 0:
                    print(f"{count} images resized")
        except Exception as e:
            print(f"Error processing {filename}: {e}")

print("All images resized successfully.")
```

```python
import pandas as pd
import os
import shutil

# Define input directory containing the images
input_directory = r'C:\Users\Harsh\Downloads\Compressed\train\comp'

# Define output directory to store selected images
output_directory = os.path.join(input_directory, 'selected')

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Load the Excel sheet
excel_file = r'C:\Users\Harsh\Downloads\Compressed\TaibLabelsSorted.xlsx'  # Provide the path to your Excel file
df = pd.read_excel(excel_file)

# Iterate through each row in the Excel sheet
for index, row in df.iterrows():
    # Extract the image name from the 'image' column
    image_name = row['image']
    # Construct the paths to the image files with different extensions
    image_path_png = os.path.join(input_directory, image_name + '.png')
    image_path_jpg = os.path.join(input_directory, image_name + '.jpg')
    image_path_jpeg = os.path.join(input_directory, image_name + '.jpeg')
    # Check if the image file exists with different extensions
    if os.path.exists(image_path_png):
        # Copy the image to the output directory
        shutil.copy(image_path_png, output_directory)
        print(f"Copied {image_name}.png to 'selected' folder")
    elif os.path.exists(image_path_jpg):
        # Copy the image to the output directory
        shutil.copy(image_path_jpg, output_directory)
        print(f"Copied {image_name}.jpg to 'selected' folder")
    elif os.path.exists(image_path_jpeg):
        # Copy the image to the output directory
        shutil.copy(image_path_jpeg, output_directory)
        print(f"Copied {image_name}.jpeg to 'selected' folder")
    else:
        print(f"Image {image_name} not found in the directory")

print("All images copied successfully.")
```

```python
import cv2
import os

# Define input directory containing the selected images
input_directory = r'C:\Users\Harsh\Desktop\ML Project\selected'

# Define output directory to store processed images
output_directory = os.path.join(input_directory, 'processed')

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Define parameters
alpha = 4
beta = -4
gamma = 128
radius = 300

# Iterate through each selected image
for filename in os.listdir(input_directory):
    if filename.endswith(".png") or filename.endswith(".jpg") or filename.endswith(".jpeg"):
        # Read the image
        image_path = os.path.join(input_directory, filename)
        img = cv2.imread(image_path)

        # Resize the image to have the same radius (300 pixels)
        height, width, _ = img.shape
        if height > width:
            scale_factor = radius / height
        else:
            scale_factor = radius / width
        img = cv2.resize(img, None, fx=scale_factor, fy=scale_factor)

        # Convert image to grayscale
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Apply Gaussian blur
        img_blurred = cv2.GaussianBlur(img_gray, (0, 0), sigmaX=15)

        # Blend the images
        img_processed = cv2.addWeighted(img_gray, alpha, img_blurred, beta, gamma)

        # Write the processed image to the output directory
        output_path = os.path.join(output_directory, filename)
        cv2.imwrite(output_path, img_processed)

        print(f"Processed {filename} and saved to 'processed' folder")

print("All images processed successfully.")
```

```python
import os
import pandas as pd
from shutil import move

# Define input directory containing the processed images
input_directory = r'C:\Users\Harsh\Downloads\Compressed\train\processed2'

# Load the Excel sheet
excel_file = r'C:\Users\Harsh\Downloads\Compressed\trainLabels.csv'  # Provide the path to your Excel file
df = pd.read_csv(excel_file)

# Define output directory to store images based on labels
output_directory = os.path.join(input_directory, 'organized')
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Iterate through each row in the Excel sheet
for index, row in df.iterrows():
    # Extract image name and label from the row
    image_name = row['image']
    label = row['level']

    # Add .jpeg extension to image name
    image_name_with_extension = image_name + '.jpeg'

    # Create folder for label if it doesn't exist
    label_folder = os.path.join(output_directory, str(label))
    if not os.path.exists(label_folder):
        os.makedirs(label_folder)

    # Move the processed image to the label folder
    image_source = os.path.join(input_directory, image_name_with_extension)
    image_destination = os.path.join(label_folder, image_name_with_extension)
    move(image_source, image_destination)

    print(f"Moved {image_name_with_extension} to label folder {label}")

print("All images organized successfully.")
```
